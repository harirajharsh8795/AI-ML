```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid", palette="muted")
```

```python
df = pd.read_csv("../data/customers_50k.csv")  
```

```python
df.head()
```

```python
##Basic Dataset Overview
print("Dataset Shape:", df.shape)
print("\nData Types and Non-Null Counts:")
print(df.info())
print("\nStatistical Summary of Numerical Features:")
print(df.describe())
```

```python
## Check for Missing Values and Duplicates
print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nNumber of Duplicate Rows:", df.duplicated().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Purchased', data=df)
plt.title("Purchased Variable Distribution", fontsize=14)
plt.xlabel("Purchased (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
num_cols = ["Age", "Annual_Income", "Credit_Score", 
            "Travel_Frequency", "Web_Visits", 
            "Email_Engagement", "Probability_Score"]

df[num_cols].hist(figsize=(12,10), bins=20, color='#2E86C1')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
cat_cols = ["Gender", "Marital_Status", "Previous_Package"]

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"{col} Distribution", fontsize=14)
    plt.xticks(rotation=30)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols + ['Purchased']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
